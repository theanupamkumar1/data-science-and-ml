```python
import pandas as pd

df = pd.DataFrame({
    'town': ['monroe township', 'monroe township', 'monroe township', 'monroe township', 'monroe township', 'west windsor', 'west windsor', 'west windsor', 'west windsor', 'robinsville', 'robinsville', 'robinsville', 'robinsville'],
    'area': [2600, 3000, 3200, 3600, 4000, 2600, 2800, 3300, 3600, 2600, 2900, 3100, 3600],
    'price': [550000, 565000, 610000, 680000, 725000, 585000, 615000, 650000, 710000, 575000, 600000, 620000, 695000]
})
```

```python
#we will get dummies dataframes for handling text that is here monre townshi,west window wtc
dummies = pd.get_dummies(df.town)
dummies
```

```python
df2 =pd.concat([df,dummies],axis='columns')
df2
```

```python
#we will drop two colums "town" becase it has text and one column from dummies because of a "dummy trap"
#problem, it is colled problem of colinear, becuase two dummies are sufficient to fing the third one

df2.drop(['town','robinsville'], axis='columns',inplace=True)
df2
```

```python
x= df2.price
x
```

```python
df3= df2.drop('price', axis='columns')
df3
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(df3,x)
```

```python
model.predict([[3400,0,0]])
```

```python
model.predict([[3800,1,0]])
```
