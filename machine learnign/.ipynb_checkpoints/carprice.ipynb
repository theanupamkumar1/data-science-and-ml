```python
import pandas as pd
import numpy as np 

df = pd.DataFrame({
    'Car Model': ['BMW X5', 'BMW X5', 'BMW X5', 'BMW X5', 'BMW X5', 'Audi A5', 'Audi A5', 'Audi A5', 'Audi A5', 'Mercedez Benz C class', 'Mercedez Benz C class', 'Mercedez Benz C class', 'Mercedez Benz C class'],
    'Mileage': [69000, 35000, 57000, 22500, 46000, 59000, 52000, 72000, 91000, 67000, 83000, 79000, 59000],
    'Sell Price($)': [18000, 34000, 26100, 40000, 31500, 29400, 32000, 19300, 12000, 22000, 20000, 21000, 33000],
    'Age(yrs)': [6, 3, 5, 2, 4, 5, 5, 6, 8, 6, 7, 7, 5]
})
df
```

```python
import matplotlib.pyplot as plt 
%matplotlib inline
plt.subplot(1, 2, 1)
plt.scatter(df["Mileage"],df["Sell Price($)"])

plt.subplot(1, 2, 2)
plt.scatter(df["Sell Price($)"],df['Age(yrs)'])
```

```python
dummies = pd.get_dummies(df[] )
```

```python
df2 =pd.concat([df,dummies], axis='columns')
df2
```

```python
df3= df2.drop(["Car Model","Audi A5"], axis="columns")
df3
```

```python
x= df3.drop(["Sell Price($)"], axis="columns")
x
```

```python
y = df3['Sell Price($)']
y
```

```python
from sklearn.linear_model import LinearRegression

hello = LinearRegression()
```

```python
hello.fit(x,y)
```

```python
hello.predict([[2500,5,0,0]])
```

```python
hello.score(x,y)
```
