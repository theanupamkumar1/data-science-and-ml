```python
subdf = df[['satisfaction_level','average_montly_hours','promotion_last_5years','salary','Work_accident']]
subdf
```

```python
salary_dummy = pd.get_dummies(subdf.salary)
salary_dummy
```

```python
merge = pd.concat([subdf,salary_dummy],axis='columns')
merge 
```

```python
final =merge.drop(['salary'], axis='columns')
x=final
x
```

```python
y= df.left  
y
```

```python
from sklearn.model_selection import train_test_split

X_test,X_train, y_test, y_train = train_test_split(x,y,train_size=0.5)
```

```python
y_train
```

```python
from sklearn.linear_model import LogisticRegression


model = LogisticRegression()
model.fit(X_train,y_train)  
```

```python
model.predict(X_test)
```

```python
model.score(X_test,y_test)
```
